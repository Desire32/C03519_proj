```python
import torch
import torch.nn as nn
import numpy as np
```

```python
x = np.array([1.0, 0.8, 0.4]) # data
w = np.array([0.9, 0.5, 0.3]) # weights
y = 1 # prediction
lr = 0.001 # learning rate
b = -0.7 # bias
```

```python
# step 1. forward pass
z = np.dot(w, x) + b
```

```python
# step 2. activation function ( in our case ReLU)
a = np.maximum(z, 0)
```

```python
# step 3. sigmoid to (0,1)
y_hat = 1 / (1 + np.exp(-a))           # sigmoid
```

```python
# step 4. counting loss
L = -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
```

```python
# step 5. backward pass ( by chain rule)
dL_dyhat = -y / y_hat                  # производная loss
dyhat_da = y_hat * (1 - y_hat)         # производная sigmoid
da_dz = 1.0 if z > 0 else 0.0         # производная ReLU
```

```python
delta = dL_dyhat * dyhat_da * da_dz
```

```python
dl_dw = delta * x
dl_db = delta
```

```python
w = w - lr * dl_dw
b = b - lr * dl_db
```

```python
print(f"loss: {L}")
print(f"w: {w}")
print(f"b: {b}")
```
